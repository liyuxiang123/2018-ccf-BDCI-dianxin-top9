```python
import numpy as np
import pandas as pd

train_data = pd.read_csv('train/train.csv')
test_data = pd.read_csv('test/test.csv')

train_data = train_data.replace('\N', -1)
train_data.gender = train_data.gender.astype(int)
train_data.age = train_data.age.astype(int)
train_data['3_total_fee'] = train_data['3_total_fee'].astype(float)
train_data['2_total_fee'] = train_data['2_total_fee'].astype(float)
```

```python
train_data1 = train_data.loc[train_data.service_type == 1]
train_data3 = train_data.loc[train_data.service_type == 3]
train_data4 = train_data.loc[train_data.service_type == 4]
test_data1 = test_data.loc[test_data.service_type == 1]
test_data3 = test_data.loc[test_data.service_type == 3]
test_data4 = test_data.loc[test_data.service_type == 4]
```

```python
def trans1(data):
    data.current_service = data.current_service.replace(90063345, 1)
    data.current_service = data.current_service.replace(90109916, 2)
    data.current_service = data.current_service.replace(90155946, 3)
    return data

def trans4(data):
    data.current_service = data.current_service.replace(89950166, 1)
    data.current_service = data.current_service.replace(89950167, 2)
    data.current_service = data.current_service.replace(99999828, 3)
    data.current_service = data.current_service.replace(89016252, 4)
    data.current_service = data.current_service.replace(89950168, 5)
    data.current_service = data.current_service.replace(99999827, 6)
    data.current_service = data.current_service.replace(99999826, 7)
    data.current_service = data.current_service.replace(99999830, 8)
    data.current_service = data.current_service.replace(99999825, 9)
    data.current_service = data.current_service.replace(89016253, 10)
    data.current_service = data.current_service.replace(89016259, 11)
    return data
```

```python
from sklearn.cross_validation import train_test_split
import lightgbm as lgb
train_data1 = trans1(train_data1)

# train_data1.gender = train_data1.gender.astype(int)
# train_data1.age = train_data1.age.astype(int)
# train_data1['3_total_fee'] = train_data1['3_total_fee'].astype(float)
# train_data1['2_total_fee'] = train_data1['2_total_fee'].astype(float)

# print train_data.columns
feature = [u'online_time', u'1_total_fee',
       u'2_total_fee', u'3_total_fee', u'4_total_fee', u'month_traffic',
       u'many_over_bill', u'contract_type', u'contract_time',
       u'net_service', u'pay_times', u'pay_num',
       u'last_month_traffic', u'local_trafffic_month', u'local_caller_time',
       u'service1_caller_time', u'service2_caller_time', u'gender', u'age',
       u'complaint_level', u'former_complaint_num', u'former_complaint_fee',]

# 'service_type', u'is_mix_service', u'is_promise_low_consume', 

train_x, evals_x, train_y, evals_y = train_test_split(train_data1[feature], train_data1.current_service,
                                                      test_size=0.2, random_state=2018)

print train_x.shape
print("LGB test")
clf = lgb.LGBMClassifier(
    boosting_type='gbdt', num_leaves=16, reg_alpha=0.0, reg_lambda=1,
    max_depth=-1, n_estimators=10000, objective='binary',
    subsample=0.7, colsample_bytree=0.7, subsample_freq=1,
    learning_rate=0.1, min_child_weight=50, random_state=2018, n_jobs=-1
)
clf.fit(train_x, train_y, eval_set=[(train_x, train_y),(evals_x, evals_y)], 
        eval_metric='multi_error',early_stopping_rounds=500, 
        categorical_feature=['many_over_bill', 'gender'],)

res1 = train_data1.user_id
```
